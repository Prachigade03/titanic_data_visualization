```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("train.csv")

# 1. Bar plot - Survival rate by Sex
sns.barplot(data=df, x="Sex", y="Survived")
plt.title("Survival Rate by Sex")
plt.savefig("survival_by_sex.png")   #  Save plot
plt.show()

# 2. Violin plot - Age distribution by Class & Survival
sns.violinplot(data=df, x="Pclass", y="Age", hue="Survived", split=True)
plt.title("Age Distribution by Class & Survival")
plt.savefig("age_violin.png")   #  Save plot
plt.show()

# 3. Box plot - Fare distribution by Class
sns.boxplot(data=df, x="Pclass", y="Fare")
plt.title("Fare Distribution by Class")
plt.savefig("fare_boxplot.png")   #  Save plot
plt.show()

# 4. KDE plot - Age of Survivors vs Non-Survivors
sns.kdeplot(df[df["Survived"]==1]["Age"], label="Survived", shade=True)
sns.kdeplot(df[df["Survived"]==0]["Age"], label="Did not Survive", shade=True)
plt.title("KDE Plot of Age")
plt.legend()
plt.savefig("age_kde.png")   #  Save plot
plt.show()

# 5. Swarm plot - Fare vs Survival (colored by Class)
sns.swarmplot(data=df.sample(200), x="Survived", y="Fare", hue="Pclass")
plt.title("Fare vs Survival")
plt.savefig("fare_swarm.png")   #  Save plot
plt.show()
```
